# Variogram Models

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from kriging import kriging
print(f"kriging v{kriging._VERSION}")
```

```python
lag = np.linspace(0.0, 2.0, 100)
nugget = 1.0
range = 1.0
sill = 1.0
```

```python
for key, model in kriging._MODELS.items():
    variogram = model([sill, range, nugget], lag)
    plt.plot(lag, variogram, label=key)
plt.xlabel('Lag')
plt.ylabel(r'$\gamma$-variance')
plt.legend()
```
